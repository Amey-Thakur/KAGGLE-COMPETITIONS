## 1. Environment Configuration
We begin by initializing the environment and defining paths for dependency management. The primary goal is to fetch and stage the `perch-hoplite` package and its required TensorFlow 2.20 dependencies.

```python
import os
from pathlib import Path

# Workspace initialization
WHEEL_DIR = Path("wheel")
WHEEL_DIR.mkdir(exist_ok=True)

print(f"Target wheel directory: {WHEEL_DIR.absolute()}")
```

## 2. Dependency Resolution
Using `pip download`, we retrieve the specific versions of TensorFlow and auxiliary libraries required for the Perch v2 inference engine. This creates a portable 'wheelhouse' that can be used for offline installations in competition submissions.

```python
# Clear previous artifacts and download new dependency tree
!rm -rf {WHEEL_DIR}
!mkdir -p {WHEEL_DIR}

print("Starting dependency download for perch-hoplite[tf]...")
!pip download perch-hoplite[tf] -d {WHEEL_DIR} --quiet

n_wheels = len(list(WHEEL_DIR.glob("*.whl")))
print(f"Successfully staged {n_wheels} wheel files.")
```
